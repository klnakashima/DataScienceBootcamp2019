# Layers
#### The layer is the building block of neural networks, comprised of linear combination & non-linear alteration

Deep net: hidden inner layers between input and output layers, we don't know what happends in hidden layers

Building blocks of a hidden layer are hidden units or nodes

Number of nodes/units in a layers is called the layer Width

note: input and output nodes can vary in quantity from each other as well as the Width

Number of hidden layers in a deep net is called the Depth

Parameters: weights(w) and biases(b). Found by optimizing

Hyperparameters: Width, Depth, Learning rate, etc. Pre-set by us before optimizing


Why do we need Non-Linearities aka Activation Functions?

They're needed so we can break linearity and represent more complicated relationships

A result of including non-linearities is the ability to stack Layers

Stacking layers is the process of placing a layer after another in a MEANINGFUL way


If non-linearites are removed, hidden nodes become useless
ex// inputlayer x with 8 inputs -w1> 9 width hidden layer -w2> 4 node output layer
     x  *  w1  =  h  *  w2  =  y
    1x8    8x9   1x9    9x4    1x4
    
    if h = x * w1
    and y = h * w2
    then y = x * w1 * w2
    but w1(8x9) and w2(9x4) can be multiplied
    so y = x * w`
    This makes the hidden layer irrelivent and unable to stack layers

In order to have deep nets and find complex relationships through arbitrary functions, we NEED non-linearities

Output of an activation function is binary


Softmax activation function output will always sum up to 1

Makes a valid probability distribution

Often used as the activation function of output layer in classification problems

ex// image recognition: is this a cat/dog/horse/etc
